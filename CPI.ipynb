```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv('CPI.csv', index_col=0).dropna()
X_train, X_test = df[:99].drop('CPI', axis=1), df[99:-1].drop('CPI', axis=1)
y_train, y_test = df[1:100].CPI, df[100:].CPI

xgb_r = xgb.XGBRegressor(objective ='reg:squarederror', max_depth = 8, learning_rate = 0.02, n_estimators = 10000)

xgb_r.fit(X_train, y_train)

pred = xgb_r.predict(X_test)

mean_absolute_error(y_test, pred)
```

```python
df = pd.read_csv('CPI.csv', index_col=0).dropna()
X_train, X_test = df[:99].drop('CPI', axis=1), df[99:-1].drop('CPI', axis=1)
y_train, y_test = df[1:100].CPI, df[100:].CPI

xgb_r = CatBoostRegressor(iterations=5000, learning_rate=0.0015, depth=4)

xgb_r.fit(X_train, y_train)

pred = xgb_r.predict(X_test)

mean_absolute_error(y_test, pred)
```

```python
plt.bar(X_train.columns, xgb_r.feature_importances_)
plt.xticks(rotation=90)
plt.show()
```
